```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
car_registration=pd.read_csv("data/Car_registration.csv")
car_registration.head()
```

```python
car_registration.columns
```

```python
# 제대로 된 속성명 지정해주기
car_registration.columns = ['날짜','종로구','중구','용산구','성동구','광진구','동대문구',
                            '중랑구','성북구','강북구','도봉구','노원구','은평구',
                            '서대문구','마포구','양천구','강서구','구로구','금천구',
                            '영등포구','동작구','관악구','서초구','강남구','송파구','강동구']
car_registration.columns
```

```python
car_registration.head()
```

```python
car_registration.shape
```

```python
car_registration.info()
```

```python
car_registration_drop=car_registration.dropna()
car_registration_drop
```

```python
car_registration_drop.info()
```

```python
car_registration_drop.describe()
```

```python
car_registration_drop.dtypes
```

```python
# Car registration 열을 추가하고 0으로 초기화합니다.
car_registration_drop['Car registration'] = 0

# 각 행의 모든 값의 합 구하기
car_registration_drop['Car registration'] = car_registration_drop.iloc[:, 1:].sum(axis=1)
car_registration_drop
```

```python
# 이전에 있던 구 데이터 열들 삭제
car_registration_drop.drop(car_registration.columns[1:-1], axis=1, inplace=True)
car_registration_drop
```

```python
car_registration_drop.to_csv('Car_registration_cleaned.csv', index=False)
```

```python
# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(car_registration_drop['날짜'], car_registration_drop['Car registration'], marker='o', color='orange')
plt.title('Car registration Over Time')
plt.xlabel('Date')
plt.ylabel('Car registration')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# 그래프 보여주기
plt.show()
```
