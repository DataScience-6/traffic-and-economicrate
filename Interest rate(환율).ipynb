```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
interest_rate=pd.read_csv("data/new/Interest_rate.csv",encoding="utf-8")
interest_rate.head()
```

```python
interest_rate.shape
```

```python
interest_rate.info()
```

```python
interest_rate.dtypes
```

```python
#날짜 형식 변경
interest_rate['Date'] = pd.to_datetime(interest_rate['Date'])
```

```python
# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(interest_rate['Date'],interest_rate['Interest rate'], marker='o', linestyle='-')

# 그래프 제목 및 축 레이블
plt.title('Interest rate Over Time')
plt.xlabel('Date')
plt.ylabel('Interest rate')

# x 축 라벨 회전
plt.xticks(rotation=45)

# 그래프 표시
plt.grid(True)
plt.tight_layout()
plt.show()
```
