## Numeric->Categorical valueÎ°ú Î≥ÄÌôòÌïòÍ∏∞
## ÏÇ¨Ïö©Ìïú feature : economic growth rate
### 0Ïù¥Ìïò : Í≤ΩÏ†ú ÌõÑÌá¥(negative), 0~2 : Î≥∏Ï†Ñ(normal), 
### 2~4: Î≥¥ÌÜµ(weak positive),4 Ïù¥ÏÉÅ : Í∏çÏ†ïÏ†Å(positive)

```python
min_e=df['Economic growth rate'].min()
min_e
```

```python
max_e=df['Economic growth rate'].max()
max_e
```

```python
def get_growth_state(rate):
    if rate < 0:
        return 'negative'
    elif rate >= 0 and rate < 2:
        return 'normal'
    elif rate >= 2 and rate < 4:
        return 'weak positive'
    else:
        return 'positive'

df['Economic growth state'] = df['Economic growth rate'].apply(get_growth_state)

df
```

```python
# Ìï¥Îãπ ÌååÏùºÏù¥ Ïó¥Î†§ÏûàÎäî Í≤ΩÏö∞ ÌååÏùº ÎßåÎì§Í∏∞Í∞Ä Í±∞Î∂ÄÎê®
#Ï†ÄÏû•ÌïòÍ∏∞
#df.to_csv('complete.csv', index=False)
```

## Categorical Value 2->Îçî Ï¢ãÏùÄ ÏïÑÏù¥ÎîîÏñ¥ ÏûàÏúºÏãúÎ©¥ ÎßêÏîÄÌï¥Ï£ºÏÑ∏Ïöî
### Î≤ÑÏä§ Í∞ÄÍ≤© ÏÉÅÏäπ Ï†Ñ/ÌõÑ(2Í∞ÄÏßÄ) Low/High
### ÌÉùÏãú Í∞ÄÍ≤© ÏÉÅÏäπ 3Î≤à(3Í∞ÄÏßÄ) Low/Middle/High

```python
def get_bus_fee(fee):
    if fee <= 1200.0:
        return 'Low'

    else:
        return 'High'

df['Bus fee'] = df['City bus trunk line'].apply(get_bus_fee)

df
```

```python
def get_texi_state(fee):
    if fee < 0:
        return 'Low'
    elif fee >= 0 and fee < 2:
        return 'Middle'
    else:
        return 'High'

df['Texi fee'] = df['Texi'].apply(get_texi_state)

df
```

```python
trans_fee=['City bus trunk line','City bus wide area','City bus circulation',
           'City bus late at night','Town bus','Subway','Texi','Model texi']
df.drop(columns=trans_fee, inplace=True)
```

```python
df
```

```python
df_cleaned = df.dropna()
df_cleaned
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
X=df_cleaned.drop(columns=['Date','Traffic rate','Economic growth state','Bus fee','Texi fee'])
y=df_cleaned.iloc[:,6]
X
```

```python
#apply SelectKBest class to extract top 10 best features

bestfeatures=SelectKBest(score_func=chi2,k=10)
fit=bestfeatures.fit(X,y)
dfcolumns=pd.Dataframe(X.columns)
dfscores=pd.DataFrame(fit.scores_)
```

```python
#concatenate two dataframes for better visualization

featureScores=pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns=['Specs','Score'] #name the dataframe columns

print(featureScores.nlargest(10,'Score')) #print 10 best features
```
