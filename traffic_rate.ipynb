```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
traffic_rate = pd.read_excel("traffic_rate.xlsx")
traffic_rate.head()
```

```python
traffic_rate.info()
```

```python
traffic_rate_drop = traffic_rate.dropna()
traffic_rate_drop
```

```python
traffic_rate_drop.info()
```

```python
traffic_rate_drop.describe()
```

```python
traffic_rate_drop.dtypes
```

```python
traffic_rate_drop_calculated = traffic_rate_drop.groupby('Date')['Total Traffic Volume'].sum().reset_index()
traffic_rate_drop_calculated
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Date', data=traffic_rate_drop_calculated)
plt.title('Boxplot of X')
plt.show()
```

```python
traffic_rate_drop_calculated['Date'] = pd.to_datetime(traffic_rate_drop_calculated['Date'], format='%Y%m%d')
traffic_rate_drop_calculated['Date'] = traffic_rate_drop_calculated['Date'].dt.strftime('%Y-%m-%d')
traffic_rate_drop_calculated.to_excel('traffic_rate_drop_calculated.xlsx', index=False)
```
