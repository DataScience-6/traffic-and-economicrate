```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas_price=pd.read_csv("data/Gas station_oil history.csv",encoding="utf-8")
gas_price.head()
```

```python
## 판다스 데이터프레임을 넘파이 배열로 변환
np_array = gas_price.to_numpy()

# 삭제할 행의 인덱스 : 2014년도
rows_to_delete = [0,1,2]  # 특정 인덱스를 삭제합니다.

# 넘파이 배열에서 특정 행 삭제
# 한 번만 실행해야 정상 실행
filtered_array = np.delete(np_array, rows_to_delete, axis=0)
gas_price = pd.DataFrame(filtered_array, columns=gas_price.columns)

gas_price
```

```python
gas_price.to_csv('Gas_price_cleaned.csv', index=False)
```

```python
# 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(gas_price["Date"], gas_price["Regular Gasoline"], marker='o', label='Regular Gasoline', color='blue')
plt.plot(gas_price["Date"], gas_price["Diesel for automobiles"], marker='o', label='Diesel for automobiles', color='green')
plt.title('Gasoline and Diesel Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# 그래프 보여주기
plt.show()
```
